# Estimate MI with DV Bound

This notebook consists of experiments evaluating the mutual information (MI) between summary statistics and parameters using SMILE, which is a variance-reduced version of MINE (Mutual Information Neural Estimator). MINE is based on the Donsker-Varadhan lower bound. However, the results demonstrate that this lower bound exhibits high variance and instability during training. At present, we have yet to discover an effective approach to obtain reliable results using this method.

```python
from tools.Trainer import train_ba,train_mine
import numpy as np
from tools.Trainer import feature_selection
```

## Pure Signal

### 1. ST

```python
st_pure=np.load('dataset/data_for_paper/pure_signal/ST.npy')
param=np.load('dataset/data_for_paper/pure_signal/params.npy')
```

```python
st_pure=(st_pure-np.mean(st_pure,axis=0))/np.std(st_pure,axis=0)

param=(param-np.array([5,1.7]))/np.array([1,0.7])
```

```python
train_MIs,eval_MIs,total_MIs=train_mine(
    st_pure,
    param,
    MI_estimator_name='smile',
    train_info={'epochs':2000,'batch_size':100,'learning_rate':1e-4,'weigth_decay':1e-4},
    verbose=True)
```

## Plot All results

```python
re_all=[{'ST_pure': 8.20,
  'BS&PS_pure': 8.65,
  'PS_pure': 6.85,
  'BS_pure': 6.46},
 {'ST_ns': 1.46,
  'BS&PS_ns': 1.47,
  'PS_ns': 1.05,
  'BS_ns': 0.58},
 {'ST_obs': 1.01,
  'BS&PS_obs': 0.90,
  'PS_obs': 0.70,
  'BS_obs': 0.08}]
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(21,7),facecolor=(1,1,1))

tick_names=[]
mi_total=[]
for i in range(len(re_all)):
    names, mis = zip(*re_all[i].items())
    
    plt.bar(np.arange(i*4,i*4+4),mis)
    
    mi_total=mi_total+list(mis)
    tick_names=tick_names+list(names)

plt.xticks(range(len(tick_names)), labels=tick_names,fontsize=15)
for i in range(len(tick_names)):
    plt.text(i-0.3, np.array(mi_total[i])+0.05, "{:.2f}".format(mi_total[i]), color='black', fontweight='bold',fontsize=15)

mi_total=np.array(mi_total)
plt.yticks(fontsize=15)
patches=[]
patches.append(mpatches.Patch(color='tab:blue',label='Pure Signal'))
patches.append(mpatches.Patch(color='tab:orange',label='with Thermal Noise'))
patches.append(mpatches.Patch(color='tab:green',label='with Thermal Noise and Residual Foreground'))
fig.legend(handles=patches,loc=(0.62,0.75),fontsize=15)
plt.ylabel(r"Estimated MI",fontsize=30)
plt.xlabel("Summary Statistics",fontsize=30)
```

Please note that the results presented here may not be very reliable due to the presence of high variance during the training process.
